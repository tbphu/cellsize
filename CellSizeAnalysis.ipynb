```python
# remove outliers 2
for pos, d in enumerate(data_single):
    d2 = d.copy()
    for r in range(len(d2)):
        row = d2[r]
        for i in range(len(row)):
            if i==0 or np.isnan(row[i-1]):
                if np.isnan(row[i+1:i+3]).any():
                    continue
                m = np.mean(row[i+1:i+3])
            elif i==len(row)-1 or np.isnan(row[i+1]):
                if np.isnan(row[i-2:i]).any():
                    continue 
                m = np.mean(row[i-2:i])
            else:
                m = np.mean( [row[i-1], row[i+1]] )
            if abs(m-row[i]) > m*0.5:
                row[i] = np.nan
        d2[r] = row
    data_single[pos] = d2
```

```python
# alighned to first occurrence
# take only cells that have daughters
data_aligned_occ = [None] * len(data_single)
for pos, d in enumerate(data_single):
    if identities[pos] is None:
        continue
    aligned = np.zeros((len(newborn[pos]), d.shape[1]))
    aligned[:] = np.nan
    for nb_pos, nb in enumerate(newborn[pos]):
        #print pos, nb
        daughters = newborn[pos][nb][2]
        if len(daughters)==0:
            continue
        first_tracked = newborn[pos][nb][0]
        aligned[nb_pos, :(d.shape[1]-first_tracked+1) ] =  ( d[nb, (first_tracked-1):]  )
    data_aligned_occ[pos] = aligned
        
```

```python
# align to first bud
# we only consider the first bud for each mother cell, all other daughers are omitted
data_aligned_bud = [None] * len(data_single)
data_aligned_daughters = [None] * len(data_single)
mother_daughter_ids = [None] * len(data_single)

for pos, d in enumerate(data_single):
    if identities[pos] is None:
        continue
    aligned = np.zeros((len(newborn[pos]), 2*no_tp))
    aligned[:] = np.nan
    aligned_daughter = np.zeros((len(newborn[pos]), 2*no_tp))
    aligned_daughter[:] = np.nan
    mother_daughter = []
    for nb_pos, nb in enumerate(newborn[pos]):
        daughters = newborn[pos][nb][2]
        if len(daughters)==0:
            mother_daughter.append([None] * 3)
            continue
        if daughters[0] < 0:
            first_birth = -daughters[0]
        else:
            first_birth = newborn[pos][daughters[0]][0]
            aligned_daughter[nb_pos, (no_tp-first_birth):(no_tp*2-first_birth) ] = d[daughters[0], :]
        aligned[nb_pos, (no_tp-first_birth):(no_tp*2-first_birth) ] = d[nb, :]
        mother_daughter.append((nb, nb_pos, daughters[0]))
        
    data_aligned_bud[pos] = aligned
    data_aligned_daughters[pos] = aligned_daughter
    mother_daughter_ids[pos] = mother_daughter
```

```python
# the IDs of the cells used in the paper (in the 0..153 range)
# we have to do this because we want to compare these positions to
# the 
paper_orig_id = [not_nan_to_orig_pos[opos] for ppos, opos in paper_id_to_here_id]
len(paper_orig_id)
```

```python
# collect mother and daugher IDs for all 3 positions in one array (and give them unique IDs)
all_mothers = []
all_positions = []
all_daughters = []

count = -1
for i in range(3):
    mapping = mother_daughter_ids[i]
    mothers, positions, daughters = zip(*mapping)
    for pos, d in enumerate(daughters):
        count += 1
        if d is None:
            all_mothers.append(None)
            all_positions.append(None)
            all_daughters.append(None)
        else:
            all_mothers.append('%s-%s' %(i, str(mothers[pos])))
            all_positions.append('%s-%s' %(i, str(positions[pos])))
            all_daughters.append('%s-%s' %(i, str(daughters[pos])))
```

```python
# get the mother's and daughters that were used in the paper
paper_mothers = [
    x for pos, x in enumerate(all_mothers)
    if pos in paper_orig_id
]
paper_daughters = [
    x for pos, x in enumerate(all_daughters)
    if pos in paper_orig_id
]
```
